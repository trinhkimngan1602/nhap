```python
# 1. CÀI ĐẶT THƯ VIỆN
!pip install folium geopy ipywidgets requests -q

import folium
from folium import plugins
from geopy.distance import geodesic
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML
import datetime
import random
import requests

# 2. LOGIC TÍNH TOÁN CHUYÊN SÂU (PRICING ENGINE)
class TNTPricingEngine:
    def __init__(self):
        # Cấu hình giá dựa trên Uber/Xanh SM
        self.config = {
            'Bike':      {'base': 15000, 'per_km': 5000,  'per_min': 400,  'platform': 2000},
            'Bike+':     {'base': 20000, 'per_km': 6200,  'per_min': 600,  'platform': 2000},
            'Car 4':     {'base': 30000, 'per_km': 12500, 'per_min': 800,  'platform': 3000},
            'Car 4 Prem':{'base': 45000, 'per_km': 16000, 'per_min': 1200, 'platform': 3000},
            'Car 7':     {'base': 55000, 'per_km': 18500, 'per_min': 1500, 'platform': 5000}
        }

    def get_real_time_data(self, lat, lng):
        """Tự động lấy thông tin môi trường mà không cần người dùng nhập"""
        now = datetime.datetime.now()
        hour = now.hour

        # Giả lập lấy Weather từ API (Có thể thay bằng API Key thật)
        # Ở đây dùng thuật toán ngẫu nhiên có căn cứ theo mùa/giờ
        weather = 'Rain' if random.random() < 0.2 else 'Sunny' # 20% khả năng mưa
        temp = random.randint(28, 39)

        # Xác định Cao điểm
        is_rush_hour = (7 <= hour <= 9) or (16 <= hour <= 19)

        # Xác định Traffic (Nếu là giờ cao điểm -> 80% kẹt xe)
        traffic_prob = random.random()
        if is_rush_hour:
            traffic_status = 'Red' if traffic_prob > 0.3 else 'Yellow'
        else:
            traffic_status = 'Green' if traffic_prob > 0.2 else 'Yellow'

        return {
            'hour': hour,
            'weather': weather,
            'temp': temp,
            'traffic': traffic_status,
            'is_rush_hour': is_rush_hour
        }

    def calculate_fare(self, dist, duration, v_type, env_data):
        cfg = self.config[v_type]
        # Base calculation
        extra_dist = max(0, dist - 2)
        subtotal = cfg['base'] + (extra_dist * cfg['per_km']) + (duration * cfg['per_min'])

        # Multipliers
        m = 1.0
        if env_data['is_rush_hour']: m *= 1.5
        if env_data['weather'] == 'Rain': m *= 1.3
        if env_data['temp'] > 38: m *= 1.1
        if env_data['traffic'] == 'Red': m *= 1.4
        elif env_data['traffic'] == 'Yellow': m *= 1.2

        final = (subtotal * m) + cfg['platform']
        return round(final, -3), m

# 3. GIAO DIỆN APP TNT
class TNTApp:
    def __init__(self):
        self.engine = TNTPricingEngine()
        self.origin = [10.762622, 106.660172] # Mặc định Quận 1, HCM
        self.dest = None
        self.output = widgets.Output()

        # UI Components
        self.logo = widgets.HTML("""
            <div style="text-align:center; background-color:#222; padding:10px; border-radius:10px;">
                <h1 style="color:#FF4500; margin:0;">💣 TNT - TARGET NEXT TRANSPORT</h1>
                <p style="color:white; margin:0;">BOMB SPEED - SMART PRICE</p>
            </div>
        """)

        self.v_select = widgets.SelectionSlider(
            options=['Bike', 'Bike+', 'Car 4', 'Car 4 Prem', 'Car 7'],
            description='Chọn xe:', style={'description_width': 'initial'}
        )
        self.btn_confirm = widgets.Button(description='🚀 ĐẶT XE NGAY', button_style='warning')
        self.btn_confirm.on_click(self.process_booking)

    def process_booking(self, b):
        with self.output:
            clear_output()
            if not self.dest:
                print("❌ Vui lòng nhập tọa độ Điểm Đến hoặc Click trên bản đồ!")
                return

            # 1. Thu thập dữ liệu thực tế
            env = self.engine.get_real_time_data(self.dest[0], self.dest[1])
            dist = geodesic(self.origin, self.dest).km
            duration = dist * 3 # Giả lập 3 phút/km

            # 2. Tính tiền
            price, multiplier = self.engine.calculate_fare(dist, duration, self.v_select.value, env)

            # 3. Hiển thị Map mới
            m = folium.Map(location=self.origin, zoom_start=14)

            # Vẽ đường đi phân đoạn màu
            color = 'red' if env['traffic'] == 'Red' else ('yellow' if env['traffic'] == 'Yellow' else 'green')
            folium.PolyLine([self.origin, self.dest], color=color, weight=6, opacity=0.8).add_to(m)

            # Marker quả bom (Destination)
            folium.Marker(self.origin, popup="Điểm đi", icon=folium.Icon(color='blue')).add_to(m)
            folium.Marker(self.dest, popup=f"Giá: {price:,.0f} VND",
                          icon=folium.Icon(color='orange', icon='fire', prefix='fa')).add_to(m)

            # Thông tin chi tiết
            display(HTML(f"""
                <div style="border-left: 5px solid #FF4500; padding:10px; background:#f9f9f9;">
                    <b>🕒 Thời gian:</b> {env['hour']}h ({"Cao điểm" if env['is_rush_hour'] else "Bình thường"}) |
                    <b>🌦️ Thời tiết:</b> {env['weather']} ({env['temp']}°C) |
                    <b>🚦 Giao thông:</b> {env['traffic']}<br>
                    <h2 style="color:#d35400;">💰 TỔNG TIỀN: {price:,.0f} VNĐ</h2>
                    <small>Hệ số nhân đang áp dụng: x{multiplier:.2f}</small>
                </div>
            """))
            display(m)

    def display_app(self):
        # Ô nhập tọa độ giả lập cho việc click (Do Colab hạn chế click-back trực tiếp)
        self.lat_input = widgets.FloatText(description='Lat Destination:', value=10.7735)
        self.lng_input = widgets.FloatText(description='Lng Destination:', value=10.7050)

        ui = widgets.VBox([
            self.logo,
            widgets.HBox([self.lat_input, self.lng_input]),
            self.v_select,
            self.btn_confirm,
            self.output
        ])
        display(ui)

        # Cập nhật tọa độ khi nhập liệu
        def update_dest(change):
            self.dest = [self.lat_input.value, self.lng_input.value]
        self.lat_input.observe(update_dest, 'value')
        self.lng_input.observe(update_dest, 'value')
        self.dest = [self.lat_input.value, self.lng_input.value]

app = TNTApp()
app.display_app()
```
