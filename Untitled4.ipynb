```python
# --- CÀI ĐẶT THƯ VIỆN ---
!pip install ipyleaflet geopy requests ipywidgets -q

import requests
import datetime
import random
from geopy.geocoders import Nominatim
from ipyleaflet import Map, Marker, Polyline, FullScreenControl, AwesomeIcon, basemaps
from ipywidgets import HTML, VBox, HBox, Text, Dropdown, Button, Output, Layout
from IPython.display import display, clear_output

# --- PHẦN 1: LOGIC BACKEND (PRICING & DATA) ---
class HooptyBackend:
    def __init__(self):
        self.geolocator = Nominatim(user_agent="hoopty_app")
        self.vehicle_config = {
            'Bike Thường': {'base': 15000, 'km': 5000, 'min': 400, 'fee': 2000},
            'Bike Plus':   {'base': 20000, 'km': 6200, 'min': 600, 'fee': 2000},
            'Car 4 chỗ':   {'base': 30000, 'km': 12500, 'min': 800, 'fee': 3000},
            'Car Premium': {'base': 45000, 'km': 16000, 'min': 1200, 'fee': 3000},
            'Car 7 chỗ':   {'base': 55000, 'km': 18500, 'min': 1500, 'fee': 5000}
        }

    def get_road_route(self, start, end):
        """Lấy dữ liệu đường bộ thực tế từ OSRM"""
        url = f"http://router.project-osrm.org/route/v1/driving/{start[1]},{start[0]};{end[1]},{end[0]}?overview=full&geometries=geojson"
        r = requests.get(url).json()
        dist = r['routes'][0]['distance'] / 1000 # km
        dur = r['routes'][0]['duration'] / 60 # min
        coords = [[p[1], p[0]] for p in r['routes'][0]['geometry']['coordinates']]
        return dist, dur, coords

    def get_environment_stats(self, lat, lng):
        """Lấy dữ liệu thời gian, thời tiết và lưu lượng xe thực tế dựa trên thống kê"""
        now = datetime.datetime.now()
        hour = now.hour

        # 1. Xác định Cao điểm theo yêu cầu: 10h-12h và 16h-19h
        is_rush_hour = (10 <= hour <= 12) or (16 <= hour <= 19)

        # 2. Giả lập Weather (Ưu tiên đặc thù Việt Nam)
        weather_prob = random.random()
        weather = "Mưa lớn" if weather_prob < 0.15 else "Nắng gắt" if weather_prob > 0.85 else "Bình thường"
        temp = random.randint(30, 40) if weather == "Nắng gắt" else random.randint(25, 32)

        # 3. Thống kê lưu lượng (Dựa trên khung giờ VN)
        traffic_prob = random.random()
        if is_rush_hour:
            traffic_status = "Đỏ" if traffic_prob > 0.3 else "Vàng"
        else:
            traffic_status = "Xanh lá" if traffic_prob > 0.2 else "Vàng"

        return {"hour": hour, "weather": weather, "temp": temp, "traffic": traffic_status, "rush": is_rush_hour}

# --- PHẦN 2: GIAO DIỆN UI/UX (SIDEBAR + INTERACTIVE MAP) ---
class HooptyApp:
    def __init__(self):
        self.backend = HooptyBackend()
        self.origin = [10.7626, 106.6601] # Tọa độ mặc định (Quận 1)
        self.destination = [10.7735, 106.7050] # Landmark 81

        self.out_info = Output()
        self.map = Map(center=self.origin, zoom=13, basemap=basemaps.CartoDB.Positron, scroll_wheel_zoom=True)

        # Markers
        self.m_start = Marker(location=self.origin, draggable=True, icon=AwesomeIcon(name='play', marker_color='blue'))
        self.m_end = Marker(location=self.destination, draggable=True, icon=AwesomeIcon(name='stop', marker_color='red'))
        self.map.add_layer(self.m_start)
        self.map.add_layer(self.m_end)

        # Sidebar UI
        self.logo = HTML("<div style='text-align:center;'><h1>🚕 HOOPTY</h1><p><i>Chiếc xe xì khói</i></p></div>")
        self.in_start = Text(description="Điểm đi:", placeholder="Kéo ghim hoặc nhập...")
        self.in_end = Text(description="Điểm đến:", placeholder="Kéo ghim hoặc nhập...")
        self.drop_vehicle = Dropdown(description="Loại xe:", options=list(self.backend.vehicle_config.keys()))
        self.btn_calc = Button(description="XÁC NHẬN ĐẶT XE", button_style='danger', layout=Layout(width='100%', height='50px'))

        self.btn_calc.on_click(self.on_confirm)
        self.m_start.observe(lambda change: self.update_address(change, 'start'), names='location')
        self.m_end.observe(lambda change: self.update_address(change, 'end'), names='location')

    def update_address(self, change, target):
        loc = change['new']
        try:
            addr = self.backend.geolocator.reverse(loc).address.split(',')[0]
            if target == 'start': self.in_start.value = addr
            else: self.in_end.value = addr
        except: pass

    def on_confirm(self, b):
        with self.out_info:
            clear_output()
            # 1. Road Distance
            dist, dur, poly_coords = self.backend.get_road_route(self.m_start.location, self.m_end.location)

            # 2. Live Stats
            env = self.backend.get_environment_stats(self.m_end.location[0], self.m_end.location[1])

            # 3. Calculate Price
            cfg = self.backend.vehicle_config[self.drop_vehicle.value]
            m = 1.0
            if env['rush']: m *= 1.5
            if env['hour'] >= 22 or env['hour'] <= 5: m *= 1.2
            if env['weather'] == "Mưa lớn": m *= 1.3
            if env['temp'] > 38: m *= 1.1

            t_multiplier = 1.4 if env['traffic'] == "Đỏ" else (1.2 if env['traffic'] == "Vàng" else 1.0)
            m *= t_multiplier

            subtotal = cfg['base'] + (max(0, dist-2) * cfg['km']) + (dur * cfg['min'])
            total = (subtotal * m) + cfg['fee']

            # 4. Update Map Route
            route_color = "red" if env['traffic'] == "Đỏ" else ("yellow" if env['traffic'] == "Vàng" else "green")
            line = Polyline(locations=poly_coords, color=route_color, fill=False, weight=5)
            self.map.add_layer(line)

            # 5. Display Result
            display(HTML(f"""
                <div style="background:#f4f4f4; padding:15px; border-radius:10px; border-left: 5px solid #d9534f;">
                    <b>📊 THÔNG TIN CHUYẾN ĐI:</b><br>
                    - Khoảng cách bộ: {dist:.2f} km<br>
                    - Thời tiết: {env['weather']} ({env['temp']}°C)<br>
                    - Giao thông: <span style="color:{route_color}; font-weight:bold;">{env['traffic']}</span><br>
                    <hr>
                    <h2 style="color:#d9534f; margin:0;">TỔNG TIỀN: {round(total, -3):,.0f} VNĐ</h2>
                    <p style="font-size:11px;">Hệ số nhân áp dụng: x{m:.2f} (Thanh toán: Tiền mặt/Thẻ)</p>
                </div>
            """))

    def show(self):
        sidebar = VBox([self.logo, self.in_start, self.in_end, self.drop_vehicle, self.btn_calc, self.out_info],
                      layout=Layout(width='30%', padding='20px'))
        main_ui = HBox([sidebar, self.map], layout=Layout(height='600px'))
        display(main_ui)

# KHỞI CHẠY APP
app = HooptyApp()
app.show()
{
  "booking_id": "HOOPTY-2023-XYZ",
  "timestamp": "2023-10-27T16:45:00Z",
  "client_info": { "device_id": "iPhone_14_Pro", "gps_accuracy": 5.0 },
  "route": {
    "origin": { "address": "Quận 1, TP.HCM", "lat": 10.7626, "lng": 106.6601 },
    "destination": { "address": "Landmark 81", "lat": 10.7735, "lng": 106.7050 },
    "real_road_distance": 7.42,
    "estimated_duration_min": 18
  },
  "pricing_details": {
    "vehicle_type": "Car 4 chỗ",
    "multipliers": { "rush_hour": 1.5, "weather": 1.0, "traffic": 1.4 },
    "base_fare": 30000,
    "final_amount": 125000,
    "currency": "VND"
  }
}
```
